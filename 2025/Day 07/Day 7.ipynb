```python
from functools import cache
```

```python
def getPuzzleInput():
	with open("Day 7.txt") as file:
		return file.read()
```

```python
print(sum([x.count("^") for x in getPuzzleInput().split("\n")])-1)
```

```python
tachyonManifold = [[x for x in line] for line in getPuzzleInput().split("\n")]
copyOfTachyonManifold = [line for line in tachyonManifold] # to manipulate
# print(copyOfTachyonManifold)

startPosition = tachyonManifold[0].index("S")
possibleTimelines = 0

@cache
def splitBeam(row, col):
	global possibleTimelines
	# for line in tachyonManifold: print(''.join(line))
	# print("row, col:", row, col)
	# print("max:", len(tachyonManifold)-1, len(tachyonManifold[0])-1)
	# print()
	
	while row < len(tachyonManifold) and tachyonManifold[row][col] != "^":
		tachyonManifold[row][col] = "|"
		row += 1
	
	# print("row, col pt 2:", row, col)
	if row == len(tachyonManifold):
		return
	
	if col-1 > -1:
		tachyonManifold[row][col-1] = "|"
		splitBeam(row+1, col-1)
		possibleTimelines += 1
	
	if col+1 < len(tachyonManifold[0]):
		tachyonManifold[row][col+1] = "|"
		splitBeam(row+1, col+1)
		possibleTimelines += 1


splitBeam(1, startPosition)

for line in tachyonManifold: print(''.join(line))

beamsSplit = 0
for row in range(len(tachyonManifold)):
	for col in range(len(tachyonManifold[row])):
		if tachyonManifold[row][col] == "^" and tachyonManifold[row-1][col] == "|":
			beamsSplit += 1

print(f"Number of beams split: {beamsSplit}")
print(f"Possible number of timelines: {possibleTimelines}")
```

```python
tachyonManifold = [[x for x in line] for line in getPuzzleInput().split("\n")]
copyOfTachyonManifold = [line for line in tachyonManifold] # to manipulate
# print(copyOfTachyonManifold)

startPosition = tachyonManifold[0].index("S")
possibleTimelines = 0
positions = []

@cache
def splitBeam(row, col):
	global possibleTimelines, positions
	# for line in tachyonManifold: print(''.join(line))
	# print("row, col:", row, col)
	# print("max:", len(tachyonManifold)-1, len(tachyonManifold[0])-1)
	# print()
	
	while row < len(tachyonManifold) and tachyonManifold[row][col] != "^":
		tachyonManifold[row][col] = "|"
		row += 1
	
	# print("row, col pt 2:", row, col)
	if row == len(tachyonManifold):
		return
	
	if col-1 > -1:
		tachyonManifold[row][col-1] = "|"
		positions.append((row, col))
		splitBeam(row+1, col-1)
		possibleTimelines += 1
	
	if col+1 < len(tachyonManifold[0]):
		tachyonManifold[row][col+1] = "|"
		positions.append((row, col))
		splitBeam(row+1, col+1)
		possibleTimelines += 1


splitBeam(1, startPosition)

for line in tachyonManifold: print(''.join(line))

beamsSplit = 0
for row in range(len(tachyonManifold)):
	for col in range(len(tachyonManifold[row])):
		if tachyonManifold[row][col] == "^" and tachyonManifold[row-1][col] == "|":
			beamsSplit += 1

print(f"Number of beams split: {beamsSplit}")
print(f"Possible number of timelines: {possibleTimelines}")
print()
print(positions)
print(Counter(positions))
print(sum([value for value in Counter(positions).values() if value == 2]))
```
