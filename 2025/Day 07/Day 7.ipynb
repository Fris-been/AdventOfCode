```python
from functools import cache
```

```python
def getPuzzleInput():
	with open("Day 7.txt") as file:
		return file.read()
```

```python
manifold = [[x for x in line] for line in getPuzzleInput().split("\n")]
copyOfTachyonManifold = [line for line in manifold] # to manipulate

startPosition = manifold[0].index("S")

@cache
def splitBeam(row, col):
	while row < len(manifold)-1 and manifold[row][col] != "^":
		manifold[row][col] = "|"
		row += 1
	
	if row == len(manifold)-1:
		return
	
	if col-1 > -1:
		manifold[row][col-1] = "|"
		splitBeam(row+1, col-1)
	
	if col+1 < len(manifold[row]):
		manifold[row][col+1] = "|"
		splitBeam(row+1, col+1)


splitBeam(1, startPosition)

for line in manifold: print(''.join(line))

beamsSplit = 0
for row in range(len(manifold)):
	for col in range(len(manifold[row])):
		if manifold[row][col] == "^" and manifold[row-1][col] == "|":
			beamsSplit += 1

print(f"Number of beams split: {beamsSplit}")
```

```python
manifold = [[x for x in line] for line in getPuzzleInput().split("\n")]

for row in range(len(manifold)):
	for col in range(len(manifold[row])):
		if manifold[row][col] == ".":
			manifold[row][col] = 0

for row in range(1, len(manifold)):
	for col in range(len(manifold[row])):
		if manifold[row][col] == "^":
			manifold[row][col-1] += 1
			manifold[row][col+1] += 1

		elif str(manifold[row-1][col]).isdigit() and manifold[row-1][col] > 0:
			manifold[row][col] = manifold[row-1][col]

for line in manifold: print(''.join(list(map(str, line))))
```
