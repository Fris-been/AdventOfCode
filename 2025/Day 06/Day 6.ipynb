```python
import re
from math import prod
import numpy as np
```

```python
def getPuzzleInput():
	with open("Day 6.txt") as file:
		return file.read()
```

```python
allProblems = np.array(getPuzzleInput().split("\n"))
allProblems = [np.array(re.split(r'\s+', line)) for line in allProblems]
allProblems = np.array(allProblems, dtype=object)

totalSum = 0
for col in range(len(allProblems[0])):
	columnToComplete = list(map(int, [allProblems[i][col] for i in range(len(allProblems)-1)]))
	if allProblems[-1][col] == "+":
		totalSum += sum(columnToComplete)
	else:
		totalSum += prod(columnToComplete)

print(f"Total sum of all problems: {totalSum}")
```

```python
allProblems = [x for x in getPuzzleInput().split("\n")]

currentSign = allProblems[-1][0]
currentPool = []
totalSum = 0

def appendToPool(col):
	toAppend = []
	for i in range(len(allProblems)-1):
		if allProblems[i][col].isspace():
			continue
		else:
			toAppend.append(allProblems[i][col])
	return int(''.join(toAppend))


def addToTotal(pool, sign):
	currentSum = 0
	# first calculate the old pool, reset everything, and append the current column
	if currentSign == "+":
		currentSum += sum(currentPool)
	else:
		currentSum += prod(currentPool)
	# print("total:", totalSum)
	return currentSum


for col in range(len(allProblems[0])):
	# if it's a completely blank line we can't do any arithmetic
	if all([allProblems[i][col].isspace() for i in range(len(allProblems))]):
		totalSum += addToTotal(currentPool, currentSign)
		currentPool = []
		currentSign = allProblems[-1][col+1]
	
	# if we've reached the end, calculate the last column in the last pool and add to total
	elif col + 1 == len(allProblems[0]):
		currentPool.append(appendToPool(col))
		totalSum += addToTotal(currentPool, currentSign)

	# otherwise keep adding columns to the current pool to calculate
	else:
		currentPool.append(appendToPool(col))
	
print(f"Total sum of all problems: {totalSum}")
```
