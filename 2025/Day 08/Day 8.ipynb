```python
def getPuzzleInput():
	with open("Day 8.txt") as file:
		return file.read()

def getDistance(point1, point2):
	from math import sqrt
	
	xs = (point2[0] - point1[0])**2
	ys = (point2[1] - point1[1])**2
	zs = (point2[2] - point1[2])**2
	return sqrt(xs + ys + zs)
```

```python
listOfCoords = getPuzzleInput().split("\n")

allDistances = []

# create every possible distance between each point
for i in range(len(listOfCoords)):
	base = tuple(map(int, listOfCoords[i].split(",")))
	for j in range(len(listOfCoords)):
		if i == j:
			continue
		
		nextPoint = tuple(map(int, listOfCoords[j].split(",")))
		allDistances.append((base, nextPoint, getDistance(base, nextPoint)))

# sort and get rid of duplicates so we can easily go through points from closest to furthest
allDistances.sort(key=lambda x: x[2])
allDistances = [a for i, a in enumerate(allDistances) if i % 2 == 0]

circuits = []

i = 0
while i < 1000:
	index1 = -1
	index2 = -1

	for circuit in range(len(circuits)):
		if allDistances[i][0] in circuits[circuit]:
			index1 = circuit
		if allDistances[i][1] in circuits[circuit]:
			index2 = circuit
	
	# i.e. neither is part of any circuit
	if index1 < 0 and index2 < 0:
		# start a new circuit with these two points
		circuits.append([allDistances[i][0], allDistances[i][1]])
	# i.e. the first point is part of a circuit and we're adding a new point
	elif index1 >= 0 and index2 == -1:
		circuits[index1].append(allDistances[i][1])

	# i.e. the second point is part of a circuit and we're adding a new point
	elif index2 >= 0 and index1 == -1:
		circuits[index2].append(allDistances[i][0])

	# if we can join two existing circuits together
	elif index1 != index2 and index1 > -1 < index2:
		circuits[index1].extend(circuits.pop(index2))
	
	i += 1
	
circuits.sort(key=lambda c:[len(c), c[0]], reverse=True)
print(f"Three largest circuits multiply to: {eval('*'.join(map(str, [len(c) for c in circuits[:3]])))}")
```

```python
circuits = []
junctionsCombined = []

i = 0
while i < len(allDistances):
	index1 = -1
	index2 = -1

	for circuit in range(len(circuits)):
		if allDistances[i][0] in circuits[circuit]:
			index1 = circuit
		if allDistances[i][1] in circuits[circuit]:
			index2 = circuit

	# neither point is part of any circuit
	if index1 < 0 and index2 < 0:
		# start a new circuit with these two points
		circuits.append([allDistances[i][0], allDistances[i][1]])

	# the first point is part of a circuit and we're adding a new point
	elif index1 >= 0 and index2 == -1:
		circuits[index1].append(allDistances[i][1])

	# the second point is part of a circuit and we're adding a new point
	elif index2 >= 0 and index1 == -1:
		circuits[index2].append(allDistances[i][0])

	# if we can join two existing circuits together
	elif index1 != index2 and index1 > -1 < index2:
		circuits[index1].extend(circuits.pop(index2))
		junctionsCombined.append(allDistances[i])
	
	i += 1

print(f"X distance between last junctions combined: {junctionsCombined[-1][0][0] * junctionsCombined[-1][1][0]}")
```
